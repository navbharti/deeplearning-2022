```
# before execution of this cell this path of image "mypic.png"
img_small = cv2.imread('/content/mypic.png', cv2.IMREAD_GRAYSCALE)
img = cv2.imread('/content/mypic.png', cv2.IMREAD_GRAYSCALE)
#cv2_imshow(img_small);
cv2_imshow(img);
#px = img[100,100]
np.set_printoptions(threshold=np.inf, linewidth=420)
#print (img_small)
```

```
kernel = np.array([[1, 0, -1],
                   [1, 0, -1],
                   [1, 0, -1]])
dst = -signal.convolve2d(img_small, kernel, mode='valid')
#cv2_imshow(dst)
print(dst)
```

```
kernel = np.array([[1, 0, -1],
                   [1, 0, -1],
                   [1, 0, -1]])
dst = cv2.filter2D(img,-1,kernel)
cv2_imshow(dst)
#print(dst)
```

```
kernel = np.array([[1, 1, 1],
                   [0, 0, 0],
                   [-1, -1, -1]])
dst = cv2.filter2D(img,-1,kernel)
cv2_imshow(dst)
#print(dst)
```

```
kernel = np.array([[1, 1, 0],
                   [1, 0, -1],
                   [0, -1, -1]])
dst = cv2.filter2D(img,-1,kernel)
cv2_imshow(dst)
#print(dst)
```

```
kernel = np.array([[-1, -1, 0],
                   [-1, 0, 1],
                   [0, 1, 1]])
dst = cv2.filter2D(img,-1,kernel)
cv2_imshow(dst)
#print(dst)
```
